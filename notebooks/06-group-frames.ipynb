```python
import pandas as pd
import re
from pathlib import Path

project_dir = Path('..').resolve()
frames_dir = Path('/Users/andretelfer/Downloads/extracted_frames/20230627/')
```

```python
df = pd.read_csv(project_dir / 'data/raw/treatments_20230627.csv')
df
```

```python
results = pd.DataFrame(dict(fname=[f.parts[-1] for f in frames_dir.glob('*')]))
results['mouse'] = results.fname.apply(lambda x: re.match('([mf]\d+)', x).group(1))
results['recording'] = results.fname.apply(lambda x: int(re.match('.*rec(\d+)', x).group(1)))
results

results.merge(df, how='left', on='mouse')
```
